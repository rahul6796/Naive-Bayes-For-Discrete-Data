```python
# function for calculating the p_cap_on_each categories featurs names:
def calculate_p_cap_on(data_frame,column_name):
    # first we find out the unique values on the given data_frame of its column_names and print them:
    df_unique_value=data[column_name].unique()
    #print(df_unique_value)
    
    # we are going to count the frequency of each unique values which are containg the columns of the data_frame:
    p_cap_on=dict(data_frame[column_name].value_counts())
    #print(p_cap_on)
    
    # we are store the each frequency of descreat random variable of the each column of the data_fream:
    value_counts=np.array(list(p_cap_on.values()))
    #print(value_counts)
    
    #now we are going calculate relative frequecy of  each unique values in each column of Data_fream:
    rf=value_counts/data_frame.shape[0]
    #print(list(rf))
    
    # now make the dictonary of each unique values and its probablity:
    p_cap_on=dict(zip(p_cap_on.keys(),rf))
   # print(p_cap_on)
    
    # Now we are going to calculate the final p_cap_on each unique values of the column in data_frames:
    # first iterate the each unique vakues of the given column of the data_frame:
    for value in df_unique_value:
        # Now we are going to iterate the values which are not present in the column of the data_fraems:
        if value not in p_cap_on.keys():
            # we are calculating the probabilty using laplace Smoothing:
            p_cap_on[value]=1/(data_frame.shape[0]+df_unique_value.shape[0])
    # And Finally return the p_cap_on:
    return p_cap_on
```

```python
p_cap_on_e={}
p_cap_on_p={}

for column_name in data.columns[1:]:
    # calculating the p_cap_on_p:
    p_cap_on_p[column_name]=calculate_p_cap_on(p_training_data,column_name)
    # calculate the p_cap_on_e:
    p_cap_on_e[column_name]=calculate_p_cap_on(e_training_data,column_name)
```

```python
# function for categories predition:
def categories_predict(x):
    # first of all we strore the data into dict:
    x=dict(x)
    # now we seperate the all keys from the data dict:
    # store the all column into a list:
    x_keys=list(x.keys())
    # we are going to initalized the p_cap_on_e and p_cap_on_p with zeros values:
    p_cap_on_e_arr=np.zeros((1,len(x_keys)))
    p_cap_on_p_arr=np.zeros((1,len(x_keys)))
    # now travers the each column name and iniantlized its value from the smoothing probability:
    for i in range(len(x_keys)):
        # now going to initalized p_cap_on_e into p_cap_on_e_arr:
        p_cap_on_e_arr=p_cap_on_e[x_keys[i]][x[x_keys[i]]]
        # and same for the intializwd p_cap_on_p into p_cap_on_p_arr:
        p_cap_on_p_arr=p_cap_on_p[x_keys[i]][x[x_keys[i]]]
    # Now we are going to calculate the likelihood probabilty for both 'e' and 'p':
    likelihood_e=np.prod(p_cap_on_e_arr)
    likelihood_p=np.prod(p_cap_on_p_arr)
    # now calulate the posterior probabilty:
    numerator=likelihood_p
    denominator=numerator+(likelihood_e)
    posterior_p=numerator/denominator
    return int(posterior_p>0.5)
        
```

```python
# Now going to predict on validation data:
predict_class_cv=[]
for i in range(validation_data.shape[0]):
    predict_class_cv.append(categories_predict(validation_data.iloc[i,1:]))
```
